# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

def Relu(x):
    return x * (x > 0)

def dRelu(x):
    return 1. * (x > 0)

x = plt.linspace(-10,10,100)

plt.plot(x, Relu(x), 'r', label='linspace(-10,10,10)')

plt.grid()

plt.title('Relu Function')

plt.text(-7.5, 9, r'$f(x) = x * (x > 0)$', fontsize=15)

plt.show()
```

```python
plt.plot(x, dRelu(x), 'b', label='linspace(-10,10,10)')

plt.grid()

plt.title('dRelu Function')

plt.text(-10, 0.9, r'$f(x) = 1. * (x > 0)$', fontsize=15)

plt.show()
```
